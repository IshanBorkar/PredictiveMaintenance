```python
# Reshaping of the rows in the sorted format
def clip(x):
    row_shape = x.shape
    if len(row_shape) < 2:
        x = x.reshape(-1,2560)
    for i in range(len(x)):
        d = x[i]
        sort_d = np.sort(d)
        n = int(2560*0.005)
        mi = sort_d[n] # minimum value
        ma = sort_d[-n] #maximum value
        d = np.clip(d,mi,ma)

        x[i] = d
    if len(row_shape) < 2:
        x = x.reshape(-1)
    return x

# Function to find kurtosis
def get_kurtosis(x):
    x = x.reshape(-1,2560)
    kurtosis = np.array( [scipy.stats.kurtosis(i) for i in x] )
#     skew = np.array( [scipy.stats.skew(i) for i in x] )
    rms = np.array( [np.mean(i**2)**0.5 for i in x] )

    return kurtosis, rms

# Function to calculate First Predicting Time
def get_First_Predicting_Time(x):
    kurtosis, rms = get_kurtosis(x)
    k = 100
#     sample_kurt = np.sort(kurtosis[:k])[1:-1]
#     sample_rms = np.sort(rms[:k])[1:-1]
    sample_kurtosis = np.sort(kurtosis[:k])
    sample_rms = np.sort(rms[:k])
    
    minimum_kurtosis = np.mean( sample_kurtosis ) - 3*np.std( sample_kurtosis )
    maximum_kurtosis = np.mean( sample_kurtosis ) + 3*np.std( sample_kurtosis )
    k = 1
    minimum_rms = np.mean( sample_rms ) - k*np.std( sample_rms )
    maximum_rms = np.mean( sample_rms ) + k*np.std( sample_rms )
    length = 5
    First_Predicting_Time = 0
    for i in range(len(kurtosis)): # i= 0/1 
        if (kurtosis[i]<minimum_kurtosis or kurtosis[i]>maximum_kurtosis) and (rms[i]<minimum_rms or rms[i]>maximum_rms) and kurtosis[i]>2:
            for j in range(i+1,i+length+1): # j= 1/2 >>> 20/21 
                if j>=len(kurtosis): break
                if (kurtosis[j]<minimum_kurtosis or kurtosis[j]>maximum_kurtosis) and (rms[j]<minimum_rms or rms[j]>maximum_rms) and kurtosis[i]>2:
                    continue
                else:
                    break
            if j==i+length:
                First_Predicting_Time = i
                break
            i = j + 1 # i= 21/22
    return First_Predicting_Time

# Function to calculate the last predicting time
def get_Last_Predicting_Time(x):
    x2 = x.reshape(-1,2560)
    print(x2.shape)
    
    length = 5
    Last_Predicting_Time = 0
    for i in range(len(x2)): # i= 0/1 
        if np.max(x2[i])>20 or np.min(x2[i])<-20:
            for j in range(i+1,i+length+1): # j= 1/2 >>> 20/21 
                if j>=len(kurtosis): break
                if np.max(x2[i])>20 or np.min(x2[i])<-20:
                    continue
                else:
                    break
            if j==i+length:
                Last_Predicting_Time = i
                break
            i = j + 1 # i= 21/22
    return Last_Predicting_Time

# Plotting all the horizontal signals, clipped horizontal signals, kurtosis, RMS
for f in bearing_files[:]:
    if '1_1' in f or  '1_3' in f or  '1_4' in f or  '1_2' in f :
        
        h,v = load_mat_tables_dataset(f)
        plt.figure(figsize=(20,10))
        plt.subplot(311)
        plt.plot(h,'b')
        # plt.plot(np.ones(shape=(100*2560,))*3,'k')
        plt.xlabel('Data Points: Time (s)')
        plt.ylabel(f[-14:-4])
        plt.title('Horizontal Signal (h)')

        h = clip(h)
        h = np.convolve(a=h,v=[0.2,0.6,0.2],mode='same') # Clipped_Horizontal_Signal
        kurtosis, rms   = get_kurtosis(h)
        plt.figure(figsize=(20,10))
        plt.subplot(311)
        plt.plot(h,'b')
        # plt.plot(np.ones(shape=(100*2560,))*3,'k')
        # plt.xlabel('Data Points: Time (s)')
        plt.ylabel(f[-14:-4])
        plt.title('Clipped Horizontal Signal (clip_h)')

        plt.subplot(312)
        plt.plot(kurtosis,'r')
        k=200
        # plt.plot(np.ones(shape=(k,))*3,'k')
        # plt.plot(np.ones(shape=(k,))*np.mean(kurtosis[:k]),'m')

        First_Predicting_Time = get_First_Predicting_Time(h)
        plt.plot([First_Predicting_Time,First_Predicting_Time],[0,np.max(kurtosis)],'g--',lw=2)
        
        Last_Predicting_Time = get_Last_Predicting_Time(h)
        if Last_Predicting_Time>0:
            plt.plot([Last_Predicting_Time,Last_Predicting_Time],[0,np.max(kurtosis)],'k--',lw=2)
        
        # plt.xlabel('Data Points: Time (s)')
        plt.ylabel(f[-14:-4])
        plt.title('Kurtosis')

        plt.subplot(313)
        plt.plot(rms,'g')
        # plt.plot(np.ones(shape=(100,))*3,'k')
        # plt.xlabel('Data Points: Time (s)')
        plt.ylabel(f[-14:-4])
        plt.title('RMS')
        plt.show()
```
